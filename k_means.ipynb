```python
df.drop(columns=["RefSeq_accession_number"], inplace=True)
```

```python
df.drop(columns=["gene_name"], inplace=True)
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.isnull().sum().sum())
print(df.shape)
```

```python
threshold = len(df) * 0.5

df = df.dropna(axis=1, thresh=threshold)

print(df.shape)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

kmeans.fit(X_scaled)

print("K-Means Model Trained Successfully!")
```

```python
df["Cluster"] = kmeans.labels_

print(df.head())
```

```python
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score

print("Calinski-Harabasz Score:",
      calinski_harabasz_score(X_scaled, kmeans.labels_))

print("Davies-Bouldin Score:",
      davies_bouldin_score(X_scaled, kmeans.labels_))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=20)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_pca)
```

```python
from sklearn.metrics import (
    silhouette_score,
    calinski_harabasz_score,
    davies_bouldin_score
)

# Silhouette Score
sil_score = silhouette_score(
    X_scaled,
    kmeans.labels_,
    sample_size=2000,    # Faster for large datasets
    random_state=42
)

# Calinski-Harabasz Score
ch_score = calinski_harabasz_score(
    X_scaled,
    kmeans.labels_
)

# Davies-Bouldin Score
db_score = davies_bouldin_score(
    X_scaled,
    kmeans.labels_
)

print("========== K-Means Evaluation Metrics ==========")
print(f"Silhouette Score         : {sil_score:.4f}")
print(f"Calinski-Harabasz Score  : {ch_score:.4f}")
print(f"Davies-Bouldin Score     : {db_score:.4f}")
```

```python
cluster_predictions = kmeans.predict(X_pca)

print(cluster_predictions)
```
